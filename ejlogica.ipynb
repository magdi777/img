```python
import math
from collections import Counter

def calcular_factorial():
    num = int(input("Ingrese un número entero: "))
    print(f"El factorial de {num} es {math.factorial(num)}")

def es_primo():
    num = int(input("Ingrese un número entero: "))
    if num < 2:
        print(f"{num} no es primo")
        return
    for i in range(2, int(math.sqrt(num)) + 1):
        if num % i == 0:
            print(f"{num} no es primo")
            return
    print(f"{num} es primo")

def mayor_menor():
    cantidad = int(input("Ingrese la cantidad de números (mínimo 3): "))
    if cantidad < 3:
        print("Debe ingresar al menos 3 números")
        return
    numeros = [int(input(f"Número {i+1}: ")) for i in range(cantidad)]
    print(f"Mayor: {max(numeros)}, Menor: {min(numeros)}")

def es_palindromo():
    cadena = input("Ingrese una cadena: ").replace(" ", "").lower()
    print("Es palíndromo" if cadena == cadena[::-1] else "No es palíndromo")

def numero_mas_repetido():
    numeros = [int(input(f"Número {i+1}: ")) for i in range(7)]
    conteo = Counter(numeros)
    num_repetido = max(conteo, key=conteo.get)
    print(f"El número más repetido es {num_repetido} y aparece {conteo[num_repetido]} veces")

def es_anagrama():
    palabra1 = input("Ingrese la primera palabra: ").lower()
    palabra2 = input("Ingrese la segunda palabra: ").lower()
    print("Es un anagrama" if sorted(palabra1) == sorted(palabra2) else "No es un anagrama")

def ordenamiento_burbuja():
    numeros = list(map(int, input("Ingrese los números separados por espacio: ").split()))
    for i in range(len(numeros)):
        for j in range(0, len(numeros) - i - 1):
            if numeros[j] > numeros[j + 1]:
                numeros[j], numeros[j + 1] = numeros[j + 1], numeros[j]
    print("Lista ordenada:", numeros)

# Para probar las funciones, descomentar y ejecutar
# calcular_factorial()
# es_primo()
# mayor_menor()
# es_palindromo()
# numero_mas_repetido()
# es_anagrama()
# ordenamiento_burbuja()
```
